```python
import numpy as np
import cv2 as cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("ht.jpeg")
numpyImg = img+100
plt.imshow(numpyImg[:,:,::-1])
# Jika Anda ingin melakukan penambahan dengan kontrol langsung dan mungkin melewati batas nilai maksimum, Anda dapat menggunakan operator NumPy.
```

```python
img.shape
nparr = np.ones((180,320,3),dtype=np.uint8) * 100
opencvImg = cv2.add(img,nparr)
plt.imshow(opencvImg[:,:,::-1])
print(opencvImg.shape[0])
print(opencvImg.shape[1])
print(opencvImg.shape[2])
# jika sudah mengetahui besar dab tinggi dari sebuah gambar pakai ini
```
